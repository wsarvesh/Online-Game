```python
import random as r
```

```python
files = ['animalplace_list.txt','object_list.txt','action_list.txt','food_list.txt']
```

```python
lists = {}

for i in files:
    temp = []
    with open(i,'r') as f:
        for line in f:
            l = line[:-1]
            l2 = l.lower()
            temp.append(l2)
    t1 = list(set(temp))
    t = r.sample(t1,300)
    print(i)
    print(len(t),t)
    print()
    print()
    lists[i] = t
```

```python
allplay = []
lists_new = {}
for i in files:
    r.shuffle(lists[i])
    lists_new[i] = lists[i][:240]
    allplay = allplay + lists[i][-60:]

lists_new['allplay.txt'] = allplay
files_new = files[:]
files_new.append('allplay.txt')

for i in files_new:
    print(i)
    print(len(lists_new[i]),lists_new[i])
    print()
    print()
```

```python
col = ['title','object','action','food','allplay']

data = []
data.append(col)

for i in range(240):
    l = []
    for j in files_new:
        l.append(lists_new[j][i])
    data.append(l)
    
print(len(data),data)
```

```python
import csv

csv.register_dialect('myDialect',
quoting=csv.QUOTE_ALL,
skipinitialspace=True,
lineterminator = '\n')


with open('Pictionary_Data.csv', 'w') as f:
    writer = csv.writer(f, dialect='myDialect')
    for row in data:
        writer.writerow(row)
        
f.close()
```
