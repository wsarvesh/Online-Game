# Places

```python
wrd = []

with open('placesc.txt','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            wrd.append(word)
# print(wrd)
```

```python
temp = []

for i in range(0,len(wrd)):
    if wrd[i] == "alt=":
        i = i + 2;
        l = wrd[i]
        s = " "
        i = i + 1
        while wrd[i] != '"':
#             print(l,wrd[i])
            l = l + s +wrd[i]
            i = i + 1
        temp.append(l)
```

```python
temp
```

```python
pl = []
for i in range(len(temp)):
    if temp[i] == 'List Challenges':
        i = i + 1
        while i + 6 < len(temp) and temp[i + 6] != 'List Challenges':
            pl.append(temp[i])
            i = i + 1

pl = pl[:-3]
places = []
# print(places)
for i in pl:
    x = i.split(", ")
    places.append(x[0])

places
```

# Animals

```python
animal = []

with open('animal.txt','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            animal.append(word)
    
print(len(animal),animal)
```

# Objects

```python
obj = []

with open('objevt.txt','r') as f:
    for line in f:
#         print(line)
        obj.append(''.join(list(line)[:-1]))

o = [i for i in obj if i not in animal]
objects = [i for i in o if len(i.split(" "))<=3]
print(len(obj),obj,"\n",len(o),o,objects,len(objects))
```

# Movies

```python
wrd = []

with open('movie.txt','r') as f:
    for line in f:
#         print(line)
        for word in line.split():
#             print(word)
            wrd.append(word)
```

```python
temp = []

for i in range(0,len(wrd)):
    if wrd[i] == "alt=":
        i = i + 2;
        l = wrd[i]
        s = " "
        i = i + 1
        while wrd[i] != '"':
#             print(l,wrd[i])
            l = l + s +wrd[i]
            i = i + 1
        temp.append(l)
```

```python
movie = []
for i in range(len(temp)):
    if temp[i] == 'List Challenges':
        i = i + 1
        while i + 6 < len(temp) and temp[i + 6] != 'List Challenges':
            movie.append(temp[i])
            i = i + 1

movie = movie[:-3]
movie = [i[:-7] for i in movie]
movie
```

# All Play

```python
import random
all_play = []
k = 0

for i in range(80):
    print(i)
    t = random.choice(objects)
    all_play.append(t)
    objects.remove(t)
    k = k + 1
    
    t = random.choice(movie)
    all_play.append(t)
    movie.remove(t)
    k = k + 1
    
    t = random.choice(actions)
    all_play.append(t)
    actions.remove(t)
    k = k + 1
    
    if i >= 40:
        continue
    else :
        t = random.choice(persons)
        all_play.append(t)
        persons.remove(t)
        k = k + 1
    
    if i >= 20:
        continue
    else:
        t = random.choice(places)
        all_play.append(t)
        places.remove(t)
        k = k + 1
        
        t = random.choice(animal)
        all_play.append(t)
        animal.remove(t)
        k = k + 1
    

# p = [all_play.append(random.choice(persons)) for i in range(40)]
# p = [all_play.append(random.choice(places)) for i in range(20)]
# p = [all_play.append(random.choice(animal)) for i in range(20)]

    
random.shuffle(all_play)

print(len(all_play),all_play,"\n")

# animals = [i for i in animal if i not in all_play]

# print(k,animals,len(animals))

```
